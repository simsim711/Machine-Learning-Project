## Data Preprocessing & Feature Engineering

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
selected_features = [
    'zip_code', 'building_code_description', 'category_code_description',
    'total_livable_area', 'number_of_rooms', 'number_of_bathrooms', 'year_built', 'market_value'
]
data_selected = data[selected_features]
missing_values = data_selected.isnull().sum()
missing_values
```

```python
import numpy as np
numerical_imputer = SimpleImputer(strategy='median')
data_selected['number_of_rooms'] = numerical_imputer.fit_transform(data_selected[['number_of_rooms']])
categorical_imputer = SimpleImputer(strategy='most_frequent')
categorical_encoder = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = Pipeline(steps=[
    ('imputer', numerical_imputer)
])
categorical_transformer = Pipeline(steps=[
    ('imputer', categorical_imputer),
    ('onehot', categorical_encoder)
])
data_selected['years_since_built'] = 2024 - data_selected['year_built']
data_selected['sqrt_total_livable_area'] = np.sqrt(data_selected['total_livable_area'])

numerical_cols = ['sqrt_total_livable_area', 'number_of_rooms', 'number_of_bathrooms', 'years_since_built']
categorical_cols = ['zip_code', 'building_code_description', 'category_code_description']
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
data_preprocessed = preprocessor.fit_transform(data_selected)
```

```python
y = data_selected['market_value']
X = data_selected.drop(['market_value'], axis=1)
X_preprocessed = preprocessor.fit_transform(X)
X_preprocessed.shape, y.shape
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MSE': mse, 'R2': r2}

results
```

#### Based on these metrics, the Random Forest model performs the best among the experimented models with the highest R^2 Score of 0.735, indicating it explains approximately 73.5% of the variance in the house prices. It also has a relatively lower MSE compared to other models.

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}

rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_

best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)

mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

r2_best
```

###  An R^2 score of approximately 0.782 indicates that the model explains around 78.2% of the variance in house prices, which is a strong score indicating good model performance.
